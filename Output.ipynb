```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read prediction:
prediction=np.loadtxt('prediction')
```

```python
# Plot:
fig, ax=plt.subplots(figsize=(12,6), nrows=1, ncols=1)
ax.plot(prediction)

ax.set_xlabel('step', size=16)
ax.tick_params(axis='both', which='major', direction='in', labelsize=16)
idx4=[0, 1, 2, 3]
ax.set_yticks(idx4)
labels = [item.get_text() for item in ax.get_yticklabels()]
labels4=['A', 'B', 'C', 'D']
ax.set_yticklabels(labels4)

plt.show()
```
